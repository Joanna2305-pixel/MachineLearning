### Interprétation des coefficients :

*   **Coefficient positif** : Une augmentation d'une unité de cette variable entraîne une augmentation proportionnelle de la variable cible (`MedHouseVal`), toutes choses égales par ailleurs.
*   **Coefficient négatif** : Une augmentation d'une unité de cette variable entraîne une diminution proportionnelle de la variable cible (`MedHouseVal`), toutes choses égales par ailleurs.

Par exemple, si le coefficient pour `MedInc` est de `0.4`, cela signifie que pour chaque augmentation d'un millier de dollars du revenu médian du quartier, le prix médian du logement (`MedHouseVal`) augmente d'environ 40 000 $, en supposant que toutes les autres variables restent constantes.
